```python
df_object = df.select_dtypes("object")
```

```python
import seaborn as sns
```

```python
sns.countplot(df_object['workclass'])
```

```python
import matplotlib.pyplot as pit
```

```python
pit.figure(figsize=[10,5])
sns.countplot(df_object, x="workclass")
pit.xticks(rotation=90)
```

```python
df_object.columns
```

```python
pit.figure(figsize=[18,30])
for i in range(len(df_object.columns)):
  pit.subplot(5,2,i+1)
  sns.countplot(df_object[df_object.columns[i]])
  pit.title(f"CountPlot of {df_object.columns[i]}")
  pit.xticks(rotation=90)
  pit.tight_layout()
```

```python
df_int = df.select_dtypes(exclude = "object")
```

```python
df_int
```

```python
sns.distplot(df_int['age'])
```

```python
pit.figure(figsize=(18,30))
for i in range(len(df_int.columns)):
  pit.subplot(5,2,i+1)
  sns.displot(df_int[df_int.columns[i]])
  pit.title(f"displot of {df_int.columns[i]}")
  pit.xticks(rotation=90)
  pit.tight_layout()
```
